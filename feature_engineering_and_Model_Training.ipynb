```python
import pandas as pd 
```

```python
#load the dataset(the preprocessing and analysis of the data is already done, in data_preprocessing and data_manipulation)
df = pd.read_csv("/home/jellyfish/Music/Delhi-Air-Quality-Predictor-End-to-End/delhi_air_quality_feature_store_processed.csv")
```

```python
df.head()
```

```python
df.drop('location_id', axis = 1)
```

Convert the event_timestamp column is in a proper datetime format and separate it into useful components like year, month, day, hour, and minute.

```python
df['event_timestamp'] = pd.to_datetime(df['event_timestamp'])

df['year'] = df['event_timestamp'].dt.year
df['month'] = df['event_timestamp'].dt.month
df['day'] = df['event_timestamp'].dt.day
df['hour'] = df['event_timestamp'].dt.hour
df['minute'] = df['event_timestamp'].dt.minute
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encoding 'city' column
label_encoder = LabelEncoder()
df['city_encoded'] = label_encoder.fit_transform(df['city'])
```

```python
df.head()
```

```python
df = df.drop(['location_id', 'city', 'event_timestamp'], axis=1)
```

```python
df.head()
```

XGBoost doesn’t require feature scaling for most datasets, but it's good practice to scale some features, especially if they have large differences in magnitude (e.g., temperature, humidity)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['temperature', 'humidity', 'pressure', 'wind_speed', 'pm25', 'pm10', 'no2', 'so2', 'o3', 'co', 'aqi']] = scaler.fit_transform(
    df[['temperature', 'humidity', 'pressure', 'wind_speed', 'pm25', 'pm10', 'no2', 'so2', 'o3', 'co', 'aqi']])
```

```python
X = df.drop('aqi', axis=1)  
y = df['aqi']
```

```python
# Split X and y based on the day condition (1-25 for training, 26-31 for testing)
X_train = X[(X['day'] >= 1) & (X['day'] <= 25)]
X_test = X[(X['day'] >= 26) & (X['day'] <= 31)] 

# Now filter y using the same day condition to maintain alignment
y_train = y[(X['day'] >= 1) & (X['day'] <= 25)]
y_test = y[(X['day'] >= 26) & (X['day'] <= 31)]

# Verify the split sizes
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

```python
X_train.shape
```

```python
from sklearn.model_selection import train_test_split

# Splitting the training set further for early stopping
X_train_new, X_val, y_train_new, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

```python
# Combine features and target variables back together
X_train_combined = X_train.copy()
X_train_combined['aqi'] = y_train

X_test_combined = X_test.copy()
X_test_combined['aqi'] = y_test

X_val_combined = X_val.copy()
X_val_combined['aqi'] = y_val

# Save to CSV files
X_train_combined.to_csv('/home/jellyfish/Music/Delhi-Air-Quality-Predictor-End-to-End/train_set.csv', index=False)
X_test_combined.to_csv('/home/jellyfish/Music/Delhi-Air-Quality-Predictor-End-to-End/test_set.csv', index=False)
X_val_combined.to_csv('/home/jellyfish/Music/Delhi-Air-Quality-Predictor-End-to-End/val_set.csv', index=False)

print("CSV files saved successfully!")
```

```python
from xgboost import XGBRegressor
```

```python
xgb_reg = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    gamma=0.1,
    min_child_weight=3,
    objective="reg:squarederror",
    early_stopping_rounds=20,
    eval_metric="rmse",
    base_score=0.5,
    n_jobs=-1,
    random_state=42
)
```

```python
# Fit the model with a validation set
evals = [(X_train_new, y_train_new), (X_val, y_val)]
xgb_reg.fit(
    X_train_new, y_train_new,
    eval_set=evals,  
    verbose=True,
)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error, explained_variance_score, median_absolute_error
import numpy as np

# Assuming y_true and y_pred are your actual and predicted values respectively

# RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")

# MAE
mae = mean_absolute_error(y_test, y_pred)
print(f"MAE: {mae}")

# MSE
mse = mean_squared_error(y_test, y_pred)
print(f"MSE: {mse}")

# R-squared
r2 = r2_score(y_test, y_pred)
print(f"R²: {r2}")

# Adjusted R-squared
n = len(y_test)  # number of data points
p = 1  # number of predictors (change accordingly)
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print(f"Adjusted R²: {adj_r2}")

# Explained Variance
explained_variance = explained_variance_score(y_test, y_pred)
print(f"Explained Variance: {explained_variance}")

# MAPE
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"MAPE: {mape}")

# Median Absolute Error
median_ae = median_absolute_error(y_test, y_pred)
print(f"Median Absolute Error: {median_ae}")
```
